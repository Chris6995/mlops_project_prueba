```python
import sys
from pathlib import Path

# Añade src al path
sys.path.append(str(Path().resolve().parent / "src"))
```

```python
import pandas as pd
from paths import TRANSFORMED_DATA_DIR

df = pd.read_parquet(TRANSFORMED_DATA_DIR / 'tabular_data.parquet')

df
```

```python
from data_split import train_test_split
from datetime import datetime


X_train, y_train, X_test, y_test = train_test_split(
    df,
    cutoff_date=datetime(2024, 1, 25),
    target_column_name='target'
)

# Drop the 'pickup_hour' column if it exists in X_train

X_train = X_train.drop(columns=['pickup_hour', 'pickup_location_id'])
X_test = X_test.drop(columns=['pickup_hour', 'pickup_location_id'])

print(f'{X_train.shape=}')
print(f'{y_train.shape=}')
print(f'{X_test.shape=}')
print(f'{y_test.shape=}')
```

```python
# import os
# os.environ['DYLD_LIBRARY_PATH'] = '/usr/local/opt/libomp/lib'
import xgboost as xgb
# modelo Xgboost
model = xgb.XGBRegressor()
# Entrenamiento del modelo
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error

# Predicciones
y_pred_lr = model.predict(X_test)

# Evaluación
mae = mean_absolute_error(y_test, y_pred_lr)
rmse = mean_squared_error(y_test, y_pred_lr)
r2 = r2_score(y_test, y_pred_lr)

print(f"Baseline (Regresión Lineal) - MAE: {mae:.2f} | RMSE: {rmse:.2f} | R2: {r2:.2f}")
```
